#How the pixels are represented in a image

```python
def visualize_input(img, ax):              #it is kind of a heat map   #creating a fn  #instead of plt using a subplot here called as ax
 #This is the whole path
    ax.imshow(img, cmap='gray')   #there are many other cmap functions  #summe is to rep diff colors
    width, height = img.shape
    thresh = img.max()/2.5

    #below is the code to just write the pixel no on that image
    for x in range(width):
        for y in range(height):
            ax.annotate(str(round(img[x][y],1)), xy=(y,x),           #img[x][y]  decides the value to be written
                        horizontalalignment='center',
                        verticalalignment='center',
                        color='white' if img[x][y]<thresh else 'black')
```

```python
fig = plt.figure(figsize = (10,10))     #10x10 is the pixel size
ax = fig.add_subplot(111)     #add_subplot is used to add axes to a figure,(111) means 1x1 grid and 1st subplot,also written as add_subplot(1,1,1)
visualize_input(x_train[1], ax)
plt.show()                               #represents a tshirt ,it gives the normalized values between 0 and 1,0 for black and  for white....if the values are nt normalized then you will get the values frm 0 t0 255
```

#Neural Network Architecture

```python
#you have to get the best parameters...this is just the basic model
from keras.layers import Dense
from keras.layers import Flatten    #we are using flatten here to flatten the image as the image is in the form of matrix  #flattening of matrix with the help of flatten layer
model=keras.Sequential()
model.add(Flatten(input_shape=(28,28)))    #no need of AF in the input layer
model.add(Dense(392,activation='relu'))                                #28x28 =784, 784/2 =392.....so there are 392 neurons in the hidden layer
model.add(Dense(10,activation='softmax'))                               
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#one more way of plotting a graph by giving history variable
#history=model.fit(x_train,y_train,epochs=10)
```

```python
#ValueError: Shapes (32, 1) and (32, 10) are incompatible   #when we are using cce it is expecting 32 and 10 columns here but you have only one column here that is as you are using cce it expects 10 columns one for each label
#if you have categorical entropy then y_train should have 10 values instaed of being 1
#your o/p is in the form of integer so you have to convert the o/p into one hot encoding format to use cce 
#as the o/p is in integer you can use sparse cce
#we are getting this error coz......we are having 60000 training data
#by default the batch size is taken as 32
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,epochs=10)    #what is 1875?......when you divide 60000 by 1875 you get the value as 32...that is by defaukt the batch size is 32 here

#If you are taking 60000 images  one byone, for every image there are 784 neurons  in the input layer so it will take long time to converge so gradient descent comes with batch gradient descent called as mini batch gradient descent that is 32 images will be trained at a time and not one image at a time ,you can also incresae the batch size
#got a machine where you consider 32 images in one go or in one step...so 1875 is the step size that is the no of steps the model takes to train all the 60000 images with the batch size of 32 (in one step 32 images are trained and in 1875 steps 60000 image are trained)
```

Consider 32 images in one bag and there are 1875 such bags .each time  the 1875 bags are passed through the model one by one during the first epoch and it calculates its o/p then again during the 2nd epoch the 1875 bags are passed and it gives the o/p with less loss and more accuracy

```python
history=model.fit(x_train,y_train,epochs=10,validation_split=0.1)     #by default the validation split is 0 but you can take 0.1 means 10 % of the training data will be the validation data and train the model again to check the validation
```
